## 1. What is Python?


<img src="https://m.media-amazon.com/images/I/813U3FuKHzL._AC_UF1000,1000_QL80_.jpg" alt="Fun Fact" style="width: 30%; float:right;"/>

Python is a very-high level programming language that can be used for several computing tasks. These programs are written with a particular **syntax**. For example, to write or print my name on the screen, we can use the command `print('Chetan')`, written with the keyword print, the brackets (parenthesis) and the inverted commas as syntax. Python's syntax is verbose and simple, making it possible to write complex codes easily.
**Statement grouping** in Python is done using indentation and commands are separated by new lines. 

It is also an Object Oriented Programming Language (OOPS), so we can define certain Python objects in the code which takes on a set of properties making it reproducible and structured.

Python is an **interpreted** language, which is to say that a set of Python commands written together in a file to be executed as a program, or a Python script, is taken one line at a time, converted into code identifiable by our computer (machine code), and executed before going on to the next line in a top to bottom fashion. 

ðŸ’¡What then is the **difference between Unix scripts and Python scripts**? Unix Shell scripts are best at moving around files and changing text data due to their higher speeds as they interact directly with the Operating System. However, Python adds in several layers of abstractions, making it possible to write very complex scripts possible and easy. It is a very powerful language. It is also dynamically-typed, meaning that it designates the data types and structures automatically, without the need for us to tell them explicitly. All this makes Python program execution a bit slower than the same program in Unix, but far more functionally complex and diverse, yet programmatically easy.

### 2.1 Python Functions, Modules and Packages

Python in itself has several pre-defined **functions** which can be used to attain several tasks (functions are blocks of commands which take in certain inputs, do a particular task in the code and returns the output - for example, `print()` is a function which takes in the argument or input in its parenthesis and prints the given input to the screen). 

Several functions can be grouped together to provide a complete set of functions to solve a task in a particular domain. These collections of python functions in a single python file are called **modules**. Several modules taken together are called **packages**.

There are several in-built Python modules as well. However, to expand the set of functionalities, we can create and use 3rd-party packages and modules in Python, such as working on tabular data like using the Pandas package, working on numerical data and matrices using numpy, or plotting mathematical data using matplotlib.

## 6. The Power of Python

We discussed that Python can be powerful. But exactly how powerful can it be? Let us see an example where Python reads 20,000 rows of data with 97 different columns in a very short amount of time. This is done using the package Pandas. Similarly, we use a package called Matplotlib to visualise this data with ease. 

### All these topics will be discussed in detail in the coming weeks, this is just to show that Python can be used to achieve very complex tasks in very few lines of code and very swiftly

Here, we are loading positional data of 20,000 stars from a CSV table that consists of Gaia (ESA's space telescope) data for the closest stars. We are then showing all the stars as per their positional coordinates RA and Dec (Right Ascension and Declination) as follows

```python
# Importing the libraries (Pandas and Matplotlib)
import pandas as pd
import matplotlib.pyplot as plt

# Reading 20,000 Rows*97 columns of Data- Closest 20k stars from Gaia Archive (DR2)
stellar=pd.read_csv('data/closest20kstars.csv')

# Creating a matplotlib (pyplot) figure
fig = plt.figure(figsize = [10,6])

# Plotting the scatter graph with RA on x-axis and Dec on y-axis. alpha tells the opacity, s is the size
plt.scatter(stellar['ra'], stellar['dec'], alpha=0.8, s=0.7)
# Adding labels and a title for the figure
plt.xlabel('RA [$\degree$]')
plt.ylabel('Dec [$\degree$]')
plt.title('Closest Stars 20,000 stars from Gaia Archive (DR2)')
```

Let us take this a step further and make our own 3D map of the universe for the closest few stars having known radius. Here, we are taking only the stars whose radius is present in the above data. We are plotting those as points with x-axis being RA, y-axis being Dec and z-axis being the distance to the stars. The stars are also given size according to their radius, so they are all in-scale, and they are color coded based on their absolute g-band photometric magnitudes (luminosities).

We can use the magic command below to also play with the 3D graph!

```python
# Taking out only those stars whose radius is known in the data
stellar_trim=stellar[(stellar['radius_val'].notnull())]

# The magic command below can be used for interactive 3D plots. Uncomment to visualize the plot in a better way.
%matplotlib notebook

# Creating a matplotlib (pyplot) figure
fig = plt.figure(figsize = (10, 6))
# Define 3D axes
ax = plt.axes(projection ="3d")
# Plot the scatter graph using RA, Dec and distance. A colormap of Red-Yellow-Blue scale is used 
# with sizes s given by stellar radii, color c given by stellar effective temperature. Opacity is set by alpha
scatter_plot=ax.scatter3D(stellar_trim['ra'], stellar_trim['dec'], stellar_trim['s_dist'],
              s=stellar_trim['radius_val']*5, c=stellar_trim['teff_val'], cmap='RdYlBu', alpha=0.7)

#Setting labels, title and a colorbar for the figure
ax.set_xlabel('RA [$\degree$]')
ax.set_ylabel('Dec [$\degree$]')
ax.set_zlabel('Distance [pc]')
plt.title('Closest Stars with known radius')
fig.colorbar(scatter_plot, label="Effective Stellar Temprature [K]")
```
